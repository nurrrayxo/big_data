### 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
import matplotlib.pyplot as plt
import numpy as np

# Параметры квадратичной функции
a = 1
b = -2
c = 1

# Параметры гармонической функции
A = 1.5
B = 2
C = np.pi / 4  # В радианах

# Генерируем значения x для обоих графиков
x = np.linspace(-10, 10, 400)

# Вычисляем значения y для квадратичной функции
y_quadratic = a * x**2 + b * x + c

# Вычисляем значения y для гармонической функции
y_harmonic = A * np.sin(B * x + C)

# Создаем графики с помощью subplot
plt.figure(figsize=(12, 6))  # Задаем размер фигуры

# Первый подграфик для квадратичной функции
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, первый подграфик
plt.plot(x, y_quadratic, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')
plt.grid(True)
plt.legend()

# Второй подграфик для гармонической функции
plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, второй подграфик
plt.plot(x, y_harmonic, color='red', label=f'y = {A} * sin({B} * x + {C})')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')
plt.grid(True)
plt.legend()

plt.tight_layout()  # Улучшает расположение подграфиков
plt.show()  # Отображение графиков
```

### 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду

```python
import matplotlib.pyplot as plt
import numpy as np

# Параметры квадратичной функции
a = 1
b = -2
c = 1

# Параметры гармонической функции
A = 1.5
B = 2
C = np.pi / 4  # В радианах

# Генерируем значения x для обоих графиков
x = np.linspace(-10, 10, 400)

# Вычисляем значения y для квадратичной функции
y_quadratic = a * x**2 + b * x + c

# Вычисляем значения y для гармонической функции
y_harmonic = A * np.sin(B * x + C)

# Создаем график
plt.figure(figsize=(8, 6))  # Задаем размер фигуры

# График квадратичной функции с подписью и легендой
plt.plot(x, y_quadratic,color='blue', label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('Графики квадратичной и гармонической функций')
plt.grid(True)

# График гармонической функции с подписью и легендой
plt.plot(x, y_harmonic, color='red', label=f'y = {A} * sin({B} * x + {C})')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.legend(loc='upper right')  # Помещаем легенду в верхний правый угол
plt.grid(True)

plt.show()  # Отображение графиков
```

### 6.	Построить график функции, заданной параметрически,  x(t) = cos(3t),y = sin(2t),t ∈ [0.4π]

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаем массив значений параметра t в заданном диапазоне [0.4pi]
t = np.linspace(0.4 * np.pi, 2 * np.pi, 1000)

# Вычисляем соответствующие значения x(t) и y(t)
x = np.cos(3 * t)
y = np.sin(2 * t)

# Создаем график
plt.figure(figsize=(8, 6))
plt.plot(x, y, color = 'purple', label='x(t) = cos(3t), y(t) = sin(2t)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График параметрически заданной функции')
plt.grid(True)
plt.grid(color = 'y')
plt.legend()
plt.show()
```

### 7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице

```python
# Данные о годах и прожиточном минимуме
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
living_min = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

# Настройки графика
dpi = 80
fig = plt.figure(dpi=dpi, figsize=(12, 6))  # Увеличим размер графика
plt.rcParams.update({'font.size': 14})
plt.title('Средний прожиточный минимум по данным статистического комитета КР')
ax = plt.axes()
plt.xlabel("Года")
plt.ylabel("Прожиточный минимум")
ax.yaxis.grid(True, zorder=1)
xs = range(len(years))

# Построение столбчатой диаграммы
plt.bar(xs, living_min, width=0.5, color='pink', alpha=0.9, label='Прожиточный минимум')

# Настройки меток оси X и их поворот
plt.xticks(xs, years, rotation=45, ha='right')

# Отображение легенды в верхнем правом углу
plt.legend(loc='upper right')

# Отображение графика
plt.tight_layout()  # Улучшает расположение элементов на графике
plt.show()

```

### 8.	Построить круговой график по данным п. 7


```python
# Настройки графика
plt.rcParams.update({'font.size': 14})
dpi = 80
plt.figure(figsize=(8, 8))
plt.title('Распределение среднего прожиточного минимума по годам')

# Построение кругового графика
plt.pie(living_min, labels=years, startangle=140, autopct=lambda p: f'{p * sum(living_min) / 100:.2f} ', colors=plt.cm.Paired.colors)

# Легенда с указанием годов
plt.legend(
    bbox_to_anchor=(-0.16, 0.45, 0.25, 0.25),
    loc='lower left',
    labels=years
)

# Задаем равные масштабы 
plt.axis('equal')

plt.show()

 
```

### 9.	Построить график произвольной каркасной поверхности.

```python
# Создаем 3D-график
fig = plt.figure(figsize=(10, 8))
ax_3d = fig.add_subplot(projection='3d')

# Настройки меток осей
ax_3d.set_xlabel('x')
ax_3d.set_ylabel('y')
ax_3d.set_zlabel('z')

# Определяем координаты точек x и y
x = [1, 2, 3]
y = [2, 5, 6, 8]

# Создаем сетку (grid) для данных точек
xgrid, ygrid = np.meshgrid(x, y)

# Устанавливаем произвольные значения для двух точек на сетке
(xgrid[1, 2], ygrid[1, 2]) = (3, 5)
(xgrid[3, 0], ygrid[3, 0]) = (1, 8)

x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)

# Вычисляем значения z на основе x и y 
zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

# Построение поверхности на основе сетки x, y и z
ax_3d.plot_surface(xgrid, ygrid, zgrid, rstride=5, cstride=5, cmap='plasma')

plt.show()
```
