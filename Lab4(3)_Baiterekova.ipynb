## Разделение на обучающие и тестовые наборы BREAST CANCER

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_dataset['data'], breast_cancer_dataset['target'],
                                                    random_state=0)
```

```python
print("Форма массива Х_test: {}".format(X_test.shape))
print("Форма массива y_test: {}".format(y_test.shape))
```

## Исследование данных BREAST CANCER

## 	Построим модель по методу k-ближайших средних и протестируем на произвольном наборе данных

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train, y_train)
```

## Прогноз модели BREAST CANCER

```python
y_pred = knn.predict(X_test)
print("Прогноз:{}".format(X_test))
```

## Оценим качество модели DIGITS, а также сделаем вывод по качеству модели

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print(f"Точность модели: {accuracy:.2f}")
```

```python
confusion = confusion_matrix(y_test, y_pred)
print("Матрица ошибок:")
print(confusion)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred, target_names=breast_cancer_dataset.target_names)
print("Отчет по классификации:\n", report)
```
